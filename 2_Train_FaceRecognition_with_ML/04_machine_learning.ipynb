```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
```

```python
#Load numpy array
data=np.load(r'C:\Users\dell\Desktop\Minor_project_facerec\face\2_Train_FaceRecognition_with_ML\data\data_pca_50_target.npz')
data.files
```

```python
data.allow_pickle=True
```

```python
X=data['arr_0'] #pca data with 50 components
y=data['arr_1'] #target or dependent variable
```

```python
X.shape,y.shape
```

```python
X
```

```python
y
```

### split the data into tran and test

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2, stratify=y)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

### training machine learning model

```python
model_svc=SVC(probability=True)
param_grid={'C':[0.5,1,10,20,30,50],
            'kernel':['rbf','poly'],
            'gamma':[0.1,0.05,0.01,0.001,0.002,0.005],
            'coef0':[0,1]}
```

```python
model_grid=GridSearchCV(model_svc,param_grid=param_grid,scoring='accuracy',
                        cv=3,verbose=2)
```

```python
model_grid.fit(x_train,y_train)
```

```python
model_grid.best_params_
```

```python
model_final=model_grid.best_estimator_
```

```python
model_final.get_params()
```

```python
y_pred=model_final.predict(x_test)
```

```python
y_pred
```

**Classsification report**

```python
cr=metrics.classification_report(y_test,y_pred,output_dict=True)
pd.DataFrame(cr).T
```

**kappa score**

```python
metrics.cohen_kappa_score(y_test,y_pred)
```

**Area under curve(auc)**


```python
metrics.roc_auc_score(np.where(y_test=='male',1,0),np.where(y_pred=='male',1,0))
```
